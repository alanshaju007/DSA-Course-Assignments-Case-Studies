```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

#Reading the Iris DataSet.

```python
path = '/content/iris.xls'
data = pd.read_excel(path)
```

#Displaying the Iris Dataset.

```python
data.head()
```

#Calculating the mean of each column of the dataset.

```python
column_means = np.mean(data, axis=0)
print(column_means)
```

#Checking for the null values present in the dataset.

```python
null_values = data.isnull()

print(null_values)
```

#Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
plt.figure()
plt.hist(data['SL'], color='green', rwidth=0.7)
plt.title('SL')
plt.show()
```

```python
sns.pairplot(data)
```

```python
plt.scatter(data['SL'], data['SW'], s=10, c='green')
plt.title('SL vs SW')
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
sns.boxplot(x='SL', y='SW', data=data)
plt.show()
```
